## 1. Konkordans med og uten korpus

For å sjekke konkordanser i bøker og aviser benyttes kommandoen `concordance()`


### 1.1. Konkordans fra aviser

For aviser brukes attributtet `title = ` for å angi avis. Verdien kan/bør brukes med jokertegnet `%`.

```python
nb.concordance(word='homofil', corpus='avis', title = 'bergens%t%', yearfrom=1960)
```

```python
nb.concordance(word='lesbisk', corpus='avis', title = 'bergens%t%', yearfrom=1960)
```

### 1.2. Fri konkordans fra bøker

```python
nb.concordance(word='homoseksuell', corpus='bok', before=10, after=10, yearfrom=1930, yearto = 1950)
```

### 1.3.. Konkordans med stort definert korpus

Her bygger vi først et korpus. Se eksempelfil om [Korpusbygging](https://nbviewer.jupyter.org/github/DH-LAB-NB/DHLAB/blob/master/1_Bygg_korpus.ipynb) for ulike måter å definere korpus.

```python
korpus = nb.book_urn(words='homofili', ddk="306.76%", period=(1960, 2020), limit=1000)
```

Undersøk lengden på korpuset med `len()`

```python
len(korpus)
```

Vi kan dra ut et delkorpus for søkeformål med `sample()`

```python
sample(korpus, 2)
```

I konkordansen kan man sette inn `sample( <korpusnavn>, <størrelse>)`. Juster størrelse til responstiden er ok. Settes den for lavt kan man risikere å ikke få treff.

```python
nb.urn_concordance(word = 'barn', urns = sample(korpus, 50), before=8, after=8)
```

### 1.4. Subkorpus fra et gitt korpus

Definer et subkorpus med `refine_book_urns()`. Et viktig parameter er `words`, som brukes til å velge bøker som inneholder spesifikke ord. 

```python
subkorpus = nb.refine_book_urn(words = ['barn'], urns=korpus, limit=1000)
```

```python
len(subkorpus)
```

Flere ord eller ordformer kan søkes etter på en gang. 

```python
nb.urn_concordance(word = ['lesbisk', 'homofil', 'homo', 'skeiv'], urns = sample(subkorpus, 5), before = 10, after = 10)
```
