Her viser vi ulike måter man kan bygge korpus fra NBs samling. Et korpus kan enten være et utvalg bøker, eller et utvalg aviser. Avgrensningen kan være metadata eller innhold i selve teksten.

## 1. Bygg korpus fra bøker

Korpuset er en definisjon av et korpus, og består av en samling URN-er.


Korpuset defineres med metadata som dewey, emneord, navn , år, etc. Her kan [Webdewey](http://deweysearchno.pansoft.de/webdeweysearch/index.html) være til god hjelp.

For å se hvilke parametre som kan søkes på, og hva som er default-innstillinger: trykk `shift`+`tab`
inne i parentesen.

### 1.1. Bygg korpus fra metadata

```python
# Bygg et korpus fra Dewey 306.76 (Seksuell legning, transperson-identitet, interseksualitet) (trunkert)
nb.book_corpus(ddk='306.76%', limit=20)
```

```python
# Bygg et korpus der forfatteren er Vigdis Hjorth

nb.book_corpus(author='Friele%Karen%',limit=10)

# % brukes som trunkeringstegn inne i søket
```

### 1.2. Bygg korpus fra innholdsord

```python
nb.book_corpus(words='homofili', period=(1990,2000), limit=5)
```

```python
nb.book_corpus(words='lesbisk, homofil, homoseksuell, kontrær', period=(1920,2020), limit=50)
```

### 1.3. Legg korpuset i en variabel

Variabelen brukes til å referere til korpuset senere

```python
korpus = nb.book_corpus(ddk="306%", subject="homofili", period=(1960, 2020), limit=10)
korpus
```
